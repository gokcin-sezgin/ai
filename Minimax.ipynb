```python
from treelib import Node, Tree
import numpy as np
```

### OYUNUN HAMLE AĞACI OLUŞTURULDU

```python
tree= Tree()
tree.create_node("A","A")#root
def node_maker (nid,fill):
    for i in range(5):
        if fill:
            tree.create_node(tree.get_node(nid).tag+str(i),tree.get_node(nid).tag+str(i),data=np.random.randint(1000),parent=nid)
        else:
            tree.create_node(tree.get_node(nid).tag+str(i),tree.get_node(nid).tag+str(i),parent=nid)
node_maker("A",False)
for i in range (5):
    node_maker("A"+str(i),False)
    for j in range(5):
        node_maker("A"+str(i)+str(j),False)
        for k in range(5):
            node_maker("A"+str(i)+str(j)+str(k),True)
tree.show()
```

```python
tree.all_nodes()
# sadece yapraklarda data bulunuyor.
```

```python
def minimax(nid, depth):
    if  tree.get_node(nid).is_leaf():
    
        return tree.get_node(nid).data
    
    if depth%2 ==1: # maximum yapılacağı zaman (bilgisayarın oynayacağı zaman)
        best_value=-1
        for node in (tree.children(nid)):
            temp= minimax(node.identifier,tree.depth(node))
            
            best_value= np.max([best_value,temp])

        return best_value
    
    else: # oyuncunun minimumu seçeceğini düşünüyoruz
        best_value=1001
        for node in (tree.children(nid)):
            temp= minimax(node.identifier,tree.depth(node))
            best_value= np.min([best_value,temp])
            
            
        return best_value
```

```python
nodes=np.array(tree.all_nodes())
OPTIMAL_VALUE=minimax("A",0)
print("Ulaşılabilecek en büyük değer: " + str(OPTIMAL_VALUE))
for node in nodes:
    if node.data==OPTIMAL_VALUE:
        print("Kullanılan yol: "+str(node.tag))
        break # Bulunca dursun
```
